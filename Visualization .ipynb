# Histogram

```python
titanic = pd.read_csv("titanic.csv")
titanic.head(2)
```

```python
titanic.age.count(), titanic.age.value_counts()
```

```python
titanic.age.plot(kind = "bar", figsize = (12,8))
```

```python
titanic.age.value_counts(bins=10) # develops bins
```

```python
#Method-1
#Pandas base approach
titanic.age.plot(kind = "hist", figsize = (12,8), grid=1)
```

```python
#non-density
titanic.age.plot(kind = "hist", figsize = (12,8), grid=1,cumulative=True)
```

```python
# Desnity_histo
titanic.age.plot(kind = "hist", figsize = (12,8), grid=1, density=True,bins=100,cumulative=True)
```

```python
#Method-2
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize=15, ylabelsize= 20,density = True,cumulative = True)
plt.show()
```

```python
#Method-3
#Matplot-Lib
plt.figure(figsize = (10,8))
plt.hist(titanic.age.dropna(),bins = 80, density = True, cumulative= True)
plt.show()
```
