```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import file
#titanic = pd.read_csv("titanic.csv")
titanic = sns.load_dataset("titanic")
```

```python
# checking the file set-
titanic.head()
```

```python
#Example Question-1
# how many male & female on the ship (use barplot to answer)
```

```python
# answer-1-a
sns.countplot(data = titanic, x = "sex") 
plt.title("Chart-1")
plt.show()
```

```python
# answer-1-b
plt.figure(figsize=(10,8))
sns.set(font_scale=2,palette="Set2")
sns.countplot(data = titanic, x = "sex") 
plt.title("Chart-2")

plt.show()
```

```python
#Example Question-2
## how many males and females on ship of the category class1,class2 and class3 (use barplot to answer)
```

```python
# answer-2-a
sns.countplot(data = titanic, x = "sex",hue="pclass")
#hue -parameter used to add the Pclass in chart
plt.title("Chart-3")
plt.show()
```

```python
# answer-2-c
plt.figure(figsize=(10,8))
sns.set(font_scale=2)
sns.countplot(data = titanic, x = "sex",hue="pclass",palette="Set3")
plt.title("Chart-4")
plt.show()
```

```python
# answer-2-c
plt.figure(figsize=(12,8))
sns.barplot(data = titanic, x = "sex", y="survived", hue="pclass") 
plt.title("Chart-5")
plt.ylabel
plt.show()
```
